```python
def table_to_excel(table_name, engine):
    df = pd.read_sql(f'SELECT * FROM {table_name}', con=engine)
    
    datetime_columns = df.select_dtypes(include=['datetime64[ns, UTC]']).columns # Converting datetime64[ns, UTC] to datetime
    df[datetime_columns] = df[datetime_columns].apply(lambda x: x.dt.tz_convert(None))
    
    df.to_excel(f'table_{table_name}.xlsx', index=False)
    print(f'Created table_{table_name}.xlsx file!')
```

```python
tables = ['Bookings', 'Tickets', 'Ticket_flights', 'Boarding_passes', 'Flights', 'Airports', 'Aircrafts', 'Seats']

# It may take a few minutes
for table in tables:
    table_to_excel(table, engine)
```
